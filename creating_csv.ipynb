```python
import csv
from typing import Type, Dict, List
from pydantic import BaseModel, Field
```

```python
class Subtitle(BaseModel):
    sub_name: str = Field(description="Subtitle's name")
    important_info: str = Field(description="Important information")
    quantity_important_rows:int = Field(description="Quantity of important rows")
    
class FiscalYear(BaseModel):
    year: str = Field(description="Fiscal year")
    period_start: str = Field(description="Period start date")
    period_end: str = Field(description="Period end date")

class Financials(BaseModel):
    currency: str = Field(description="Give me Currency of this report")
    net_sales: float = Field(-1, description="Give me Net sales(absolute)  if you dont find it output should be -1")            
    revenue_growth_pct: float = Field(-1, description="GIve me Revenue growth percentage if you dont find it output should be -1")
    operating_profit: int = Field(-1, description="Give me Consolidated income statement as Operating profit if you dont find it output should be -1")
    operating_margin: float = Field(-1, description="Give me Operating margin percentage if you dont find it output should be -1")
    net_income_margin_pct: float = Field(-1, description="Find me Net income margin percentage if you don't find it output should be -1")
    net_profit: int = Field(-1, description="Give me Net profit if you dont find it output should be -1")
    eps: float = Field(-1, description="Give me EPS (basic/diluted) if you dont find it output should be -1")
    cash_flow:int = Field(-1, description="Give me Free cash flow / Net cash from ops if you dont find it output should be -1")
    capex:int = Field(-1, description="Give me Capex (absolute and/or cadence) if you dont find it output should be -1")
    opex: int = Field(-1, description="Give me Opex, if you dont find output should be -1")
    gross_profit: int = Field(-1, description="Give me Gross profit if you dont find it output should be -1")
    share_of_sales: float = Field(description="Find me percentage of Share of sales by region, if you don't find it output should be -1")
    gross_margin: float = Field(-1, description="Give me Gross margin percentage if you dont find it output should be -1")
    revenue: int = Field(-1, description="Give me total revenue and clearly specify the currency (EUR, USD, etc.) and unit (e.g., millions, billions). otherwise output should be -1")
    currency: str = Field("Unknown", description="Give me the Currency of this report (e.g., EUR, USD, GBP, etc.) otherwise output should be 'Unknown'")
    revenue_unit: str = Field("Unknown", description="Specify the unit for all financial amounts, e.g., 'in millions of euros', 'in billions of USD' otherwise output should be 'Unknown'")
    net_sales_unit: str = Field("Unknown", description="Specify the unit for Net Sales amount, e.g., 'in millions of euros', 'in billions of USD' otherwise output should be 'Unknown'")
    operating_income: int = Field(-1, description="Give me Operating income (ROC) if you don't find it output should be -1")
    operating_income_unit: str = Field("Unknown", description="Specify the unit for Operating Income amount, e.g., 'in millions of euros', 'in billions of USD' otherwise output should be 'Unknown'")
    net_income: int = Field(-1, description="Give me Net income (Group share) if you don't find it output should be -1")
    net_income_growth: float = Field(-1, description="Give me Net income growth percentage if you don't find it output should be -1")
    net_debt: float = Field(-1, description="Give me Net debt if you don't find it output should be -1")
    net_debt_to_ebitda: float = Field(-1, description="Give me Net debt/EBITDA ratio if you don't find it output should be -1")
    dividend: float = Field(-1, description="Give me the proposed dividend per share (e.g., €4.70) if you don't find it output should be -1")
    pro: float = Field(-1, description="Give me PRO (Profit from Recurring Operations) amount if you don't find it output should be -1")
    pro_growth: float = Field(-1, description="Give me organic PRO growth percentage if you don't find it output should be -1")
    fy24_operating_margin: float = Field(-1, description="Give me FY24 operating margin percentage if you don't find it output should be -1")
    fy24_fcf: float = Field(-1, description="Give me FY24 Free Cash Flow if you don't find it output should be -1")

class FreeCashFlow_Debt(BaseModel):
    free_cash_flow_amount: float = Field(-1, description="Free Cash Flow in €m")
    free_cash_flow_unit: str = Field("Unknown", description="Specify the unit for Free Cash Flow amount, e.g., 'in millions of euros', 'in billions of USD' otherwise output should be 'Unknown'")
    net_debt_change_amount: float = Field(-1, description="Change in Net Debt vs 30 June 2023 in €m (use sign)")
    net_debt_ending_amount: float = Field(-1, description="Net Debt at 30 June 2024 in €m")
    net_debt_to_ebitda_ratio: float = Field(-1, description="Net Debt / EBITDA ratio at average rate")
    dividend_per_share_proposed: float = Field(-1, description="Proposed dividend per share in €")
    agm_date: str = Field('Unknown',description="AGM date for dividend approval (text) otherwise 'Unknown'")
    dividend_cagr_since_fy19_pct: float = Field(-1, description="Dividend CAGR since FY19, %")
    financial_policy_unchanged: bool = Field(True, description="Whether financial policy remains unchanged")

class Region(BaseModel):
    name: str = Field("Unknown", description="Give me Region of affiliates, if you don't find it output should be 'Unknown'")
    net_sales: int = Field(-1, description="Find me Segmental Net sales  by region if you don't find it output should be -1")
    revenue: int = Field(-1, description="Find me  Revenue by region if you don't find it output should be -1")
    revenue_growth: int = Field(-1, description="Give me revenue growth by region if you don't find it output should be -1")
    yoy_pct: int = Field(-1, description="Find me Region Year over Year percentages if you don't find it output should be -1")
    currency: str = Field("Unknown",description="Currency symbol/code found for this region’s figures (e.g., '€', 'EUR', 'USD'). If not stated at region level, inherit from report. If unknown, return 'Unknown'.")
    amount_unit: str = Field("Unknown",description="Exact scale stated for amounts for this region (e.g., 'en millions', 'en milliards', 'Md€', 'M€'). Always keep the unit’s wording as written. If not stated, return 'Unknown'.")
    revenue_absolute: int = Field(-1,description="Absolute revenue for the region in the currency/scale above (extract digits only, no unit). If not given for the region, return -1.")
    revenue_absolute_unit: str = Field("Unknown", description="Unit for the absolute revenue for the region (e.g., 'millions of euros', 'billions of USD'). If not found, return 'Unknown'.")
    revenue_org_change_pct: float = Field(-1,description="Organic revenue change for the region as a signed percentage (e.g., -3.0 for -3.0%). If not found, return -1.")
    revenue_pub_change_pct: float = Field(-1, description="Published (données publiées) revenue change for the region as a signed percentage. If not found, return -1.")
    volume_growth_pct: float = Field(-1, description="Volume growth for the region as a signed percentage (e.g., +2.0) also if the text refers to Group level return -1.")

class Brands(BaseModel):
    quantity_key_brands: int = Field(description="Find me Total Quantity of Key brands from this company")
    key_brands: str = Field(description="Find me All Key brands from this company")
    brand_companies: str = Field(description="Find me All Brand Companies from this company")
    quantity_brand_companies: int = Field(-1, description="Find me Total Quantity of Brand Companies from this company otherwise output should be -1")
    strategic_international_brands: str = Field(description="Find me All Strategic International Brands from this company")
    quantity_strategic_international_brands: int = Field(-1, description="Find me Total Quantity of Strategic International Brands from this company otherwise output should be -1")
    prestige_brands: str = Field(description="Find me All Prestige Brands from this company")
    quantity_prestige_brands: int = Field(-1, description="Find me Total Quantity of Prestige Brands from this company otherwise output should be -1")
    specialty_brands: str = Field(description="Find me All Specialty Brands from this company")
    quantity_specialty_brands: int = Field(-1, description="Find me Total Quantity of Specialty Brands from this company otherwise output should be -1")
    strategic_local_brands: str = Field(description="Find me All Strategic Local Brands from this company")
    quantity_strategic_local_brands: int = Field(-1, description="Find me Total Quantity of Strategic Local Brands from this company otherwise output should be -1")
    non_alcoholic_brands: str = Field(description="Find me All Non-Alcoholic Brands from this company")
    quantity_non_alcoholic_brands: int = Field(-1, description="Find me Total Quantity of Non-Alcoholic Brands from this company otherwise output should be -1")
    ready_to_drink_brands: str = Field(description="Find me All Ready-To-Drink Brands from this company")
    quantity_ready_to_drink_brands: int = Field(-1, description="Find me Total Quantity of Ready-To-Drink Brands from this company otherwise output should be -1")

class Sales_Drinks(BaseModel):
    total_net_sales: float = Field(-1, description="Total FY24 Net Sales in million euros (absolute)")
    organic_decline_pct: float = Field(-1, description="Organic decline percentage in total net sales")
    reported_decline_pct: float = Field(-1, description="Reported decline percentage in total net sales")
    net_sales_analysis_by_period: int = Field(description="Net Sales analysis by period and region ")
    fx_impact: float = Field(-1, description="FX impact in million euros")
    perimeter_impact: float = Field(-1, description="Positive perimeter impact in million euros")
    americas_growth: float = Field(-1, description="Americas overall sales growth percentage")
    usa_growth: float = Field(-1, description="USA sales growth percentage otherwise output should be -1")
    asia_row_growth: float = Field(-1, description="Asia-Rest of World overall sales growth percentage otherwise output should be -1")
    china_growth: float = Field(-1, description="China sales growth percentage otherwise output should be -1")
    india_growth: float = Field(-1, description="India sales growth percentage otherwise output should be -1")
    spirits_market_trend_value: float = Field(-1, description="Estimated global spirits market growth or normalization percentage otherwise -1")
    spirits_market_trend: str = Field('Unknown',description="Trend summary for global spirits market normalization otherwise output should be 'Unknown'")
    inventory_adjustments: str = Field('Unknown',description="Summary of inventory adjustments mentioned in report OTHERWISE output should be 'Unknown'")
    

class Results_Drinks(BaseModel):
    pro_amount: float = Field(-1, description="FY24 Profit from Recurring Operations in €m otherwise output should be -1")
    pro_organic_growth_pct: float = Field(-1, description="Organic growth percentage of PRO otherwise output should be -1")
    pro_reported_change_pct: float = Field(-1, description="Reported change percentage of PRO otherwise output should be -1")
    gross_margin_expansion_bps: int = Field(-1, description="Organic gross margin expansion in basis points otherwise output should be -1")
    ap_amount: float = Field(-1, description="A&P spend in €bn otherwise output should be -1")
    ap_pct_of_net_sales: float = Field(-1, description="A&P as percentage of Net Sales otherwise output should be -1")
    operating_margin_org_bps: int = Field(-1, description="Operating margin organic expansion in basis points otherwise output should be -1")
    operating_margin_org_pct: float = Field(-1, description="Operating margin on an organic basis, %")
    operating_margin_reported_pct: float = Field(-1, description="Operating margin on a reported basis, %")
    fx_impact_amount: float = Field(-1, description="Adverse FX impact on reported operating margin in €m")
    perimeter_effect_amount: float = Field(-1, description="Favourable perimeter effects in €m")
    group_share_net_pro_amount: float = Field(-1, description="Group share of Net PRO in €m")
    group_share_net_pro_amount_unit: str = Field("Unknown", description="Specify the unit for Group Share of Net PRO amount, e.g., 'in millions of euros', 'in billions of USD' otherwise output should be 'Unknown'")
    group_share_net_pro_change_pct: float = Field(-1, description="YoY change percentage of Group share of Net PRO")
    avg_cost_of_debt_pct: float = Field(-1, description="Average cost of debt percentage for recurring financial expenses")
    group_share_net_profit_amount: float = Field(-1, description="Group Share of Net Profit in €m")
    group_share_net_profit_change_pct: float = Field(-1, description="YoY change percentage of Group Share of Net Profit")
    eps_amount: float = Field(-1, description="Earnings per share in €")
    
class Corporate_information(BaseModel):
    headquarters: str = Field(description="Give me Headquarters location of this company")
    executive_committee_examples: str = Field('Unknown', description="Give me all of executive committee members")
    executive_committee_quantity:int = Field(description="Give me Total quantity of executive committee members")
    board_of_directors_examples: str = Field('Unknown',description="Give me all of board of directors members otherwise output should be 'Unknown'")
    board_of_directors_quantity:int = Field(description="Give me Total quantity of board of directors members")
    affiliate_name: str = Field("Unknown",description="Exact all names of the affiliate or subsidiary company (e.g., 'Pernod Ricard USA', 'Pernod Ricard India'). If not mentioned in the report, return 'Unknown'.")
    affiliates: str = Field('Unknown', description="Give me all of names of affiliates of this company or subsidiary company (e.g., 'Pernod Ricard USA', 'Pernod Ricard India').otherwise output should be 'Unknown'")
    affiliate_quantity:int = Field(description="Give me Total quantity of affiliates of this company or subsidiary companyof this company")
    total_employees:int = Field(-1,description="Give me Total employees quantity")
    avg_age:int = Field(-1,description="Give me Average age of employees")
    qty_nationalities: int = Field(-1,description="Find me total Quantity of Nationalities if you dont find it output should be -1")
    pct_women:int = Field(-1, description="Give percentage of women in company otherwise output should be -1")
    leadership_pro: int = Field(-1, description="Give me percentage of leadership programs otherwise output should be -1")
    senior_appointments_bw: int = Field(-1, description="Give me percentage of Senior appointments by women otherwise output should be -1")
    female_leaders:int = Field(-1, description="Give me percentage of female leaders otherwise output should be -1")
    ethn_diverse_lead:int = Field(-1 ,description="Give me percentage of ethnically diverse leaders otherwise output should be -1")
    employees_with_disabilities_pct:int = Field(-1, description="Give me percentage of employees with disabilities otherwise output should be -1")
    global_mobility_pct:int = Field(-1, description="Give me percentage of global mobility otherwise output should be -1")
    nationalities:int = Field(-1, description="Give me quantity of nationalities otherwise output should be -1")

class Social_DEI(BaseModel):
    total_employees: int = Field(-1, description="Total number of employees")
    women_in_workforce_pct: float = Field(-1, description="Percentage of women in total workforce, otherwise -1")
    women_in_management_pct: float = Field(-1, description="Percentage of women in management roles, otherwise -1")
    ethnically_diverse_leaders_pct: float = Field(-1, description="Percentage of ethnically diverse leaders, otherwise -1")
    pay_gap_ratio: float = Field(-1, description="Gender or diversity pay gap ratio (men vs women), otherwise -1")
    employees_with_disabilities_pct: float = Field(-1, description="Percentage of employees with disabilities, otherwise -1")
    lgbtq_inclusion_programs: str = Field('Unknown', description="Information about LGBTQ+ inclusion or diversity initiatives otherwise output should be 'Unknown'")
    training_hours_per_employee: float = Field(-1, description="Average training hours per employee, otherwise -1")
    community_investment_amount: float = Field(-1, description="Community investments or donations in €m, otherwise -1")

class Environmental(BaseModel):
    carbon_emissions_total: float = Field(-1, description="Total carbon emissions (Scope 1+2) in metric tons CO2e, otherwise -1")
    carbon_emissions_scope3: float = Field(-1, description="Scope 3 emissions in metric tons CO2e, otherwise -1")
    emission_intensity: float = Field(-1, description="Emission intensity (tons CO2e per unit of revenue), otherwise -1")
    renewable_energy_pct: float = Field(-1, description="Percentage of renewable energy used, otherwise -1")
    energy_consumption_total: float = Field(-1, description="Total energy consumption (MWh), otherwise -1")
    water_withdrawal_total: float = Field(-1, description="Total water withdrawal in cubic meters, otherwise -1")
    waste_recycled_pct: float = Field(-1, description="Percentage of total waste recycled, otherwise -1")
    biodiversity_initiatives: str = Field('Unknown', description="Summary of biodiversity protection or environmental initiatives otherwise output should be 'Unknown'")

class Governance(BaseModel):
    water_efficiency: str = Field('Unknown', description="Find me information about Water efficiency otherwise output should be 'Unknown'")
    energy_consumption: str = Field('Unknown', description="Find me information about Energy consumption otherwise output should be 'Unknown'")
    distillery_water: str = Field('Unknown', description="Find me information about Distillery water otherwise output should be 'Unknown'")
    responsible_consumption: str = Field('Unknown', description="Find me information about Responsible consumption otherwise output should be 'Unknown'")
```

```python
class Articles(BaseModel):
    topic: str = Field(description="Find me main article of this report")
    company_name: str = Field(description="Give me Company's name")
    motto: str = Field(description="Give me Company's motto or slogan")
    founded_group: str = Field(description="Give me Founded year of this company")
    headquarters: str = Field(description="Give me Headquarters location of this company i need city and country")
    maisons_count: int = Field(description="Give me Total quantity of maisons of this company")
    countries: int = Field(description="Give me Total quantity of countries where this company is present")
    stores_count: int = Field(description="Give me Total quantity of stores of this company")
    fy: List[FiscalYear] = Field(description="Fiscal Years")
    region: List[Region] = Field(description="Regions KPTs")
    finance: List[Financials] = Field(description="Financial and capital KPTs")
    free_cash_flow_debt: List[FreeCashFlow_Debt] = Field(description="Free Cash Flow and Debt KPTs")
    sales_drinks: List[Sales_Drinks] = Field(description="Sales Drinks KPTs")
    results_drinks: List[Results_Drinks] = Field(description="Results Drinks KPTs")
    corporate: List[Corporate_information] = Field(description="Corporate information KPTs")
    social_dei: List[Social_DEI] = Field(description="Social and DEI KPTs")
    environmental: List[Environmental] = Field(description="Environmental KPTs")
    governance: List[Governance] = Field(description="Governance KPTs")
    subtitles: List[Subtitle] = Field(description="subtitles")
```

```python
group_fields: Dict[str, Type[BaseModel]] = {
    "Articles": Articles,
    "FiscalYear": FiscalYear,
    "Region": Region,
    "Financials": Financials,
    "FreeCashFlow_Debt": FreeCashFlow_Debt,
    "Brands": Brands,
    "Sales_Drinks": Sales_Drinks,
    "Results_Drinks": Results_Drinks,
    "Corporate_information": Corporate_information,
    "Social_DEI": Social_DEI,
    "Environmental": Environmental,
    "Governance": Governance,
    "Subtitle":  Subtitle
}

def build_questions_from_reports(
        reports: dict,
        models: List[str],
        out_path: str = "questions.csv",
        search_k: int = 10,
        close_context: int = 5,
):
    
    rows = []
    for item_key in reports.keys():
        for m in models:
            if m not in models:
                continue
            q_id = f"{m}_{item_key.replace('/','_')}"
            demand = f"Extract full {m} schema from this company report and answer all fields, if you don't find the information answer with -1 for numeric fields and 'Unknown' for string fields."
            rows.append({
                q_id: q_id,
                item_key: item_key,
                models: m,
                "demand": demand,
                search_k: search_k,
                close_context: close_context,
            })

    with open(out_path, mode="w", newline="", encoding="utf-8") as f:
        create_csv = csv.DictWriter(f, fieldnames=rows[0].keys())
        create_csv.writeheader()
        create_csv.writerows(rows)
build_questions_from_reports(reports, list(group_fields.keys()), out_path="questions.csv")
```
